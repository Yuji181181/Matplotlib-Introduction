```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(10)
y = np.random.randint(-10,10,10)
plt.plot(x,y)
```

```python
x1 = np.random.randint(10,20,20)
x2 = np.random.randint(20,30,20)
y1 = np.random.randint(50,100,20)
y2 = np.random.randint(0,40,20)
plt.scatter(x1,y1)
plt.scatter(x2,y2)
```

```python
x = np.arange(10)
y = np.random.randint(-10,10,10)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Result")
plt.plot(x,y)
```
