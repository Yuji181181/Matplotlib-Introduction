```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = ["A","B","C","D","E"]
y = [10,20,30,20,10]
```

```python
plt.bar(x,y)
```

```python
plt.bar(x,y,width = 0.5)
```

```python
plt.barh(x,y)
```

```python
z = ["red","blue","red","red","red"]
plt.bar(x,y,color = z)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(x,y)
```

```python
plt.bar(x,y,align="center",color="r")
plt.bar(x,y,align="edge",color="b")
```
